```python
import os, sys
import numpy as np
import pyvista as pv
import matplotlib.pyplot as plt
from scipy.constants import c 

sys.path.append('../../../')

from solverFIT3D import SolverFIT3D
from gridFIT3D import GridFIT3D 
from wakeSolver import WakeSolver



wake = WakeSolver()
wake.load_results('results_beta1_sigma1e3')
```

## Plot

```python
# CST_wake = np.loadtxt('/home/emacchia/wakis/benchmarks/betacavity/results_beta1_sigma1e3/CST_wake_z_beta1.txt')
# CST wake and impedance
cstWP = wake.read_txt('results_beta1_sigma1e3/CST_wake_z_beta1.txt')
cstZ = wake.read_txt('results_beta1_sigma1e3/CST_Z_z_beta1.txt')
cstZReIm = wake.read_txt('results_beta1_sigma1e3/CST_Z_z_ReIm_beta1.txt')


fig, ax = plt.subplots(1,2, figsize=[12,4], dpi=150)
ax[0].plot(wake.s*1e2, wake.WP, c='r', lw=1.5, label='FIT+Wakis')
ax[0].plot(cstWP[0], cstWP[1], c='k', ls='--', lw=1.5, label='CST')
ax[0].set_xlabel('s [cm]')
ax[0].set_ylabel('Longitudinal wake potential [V/pC]', color='r')
ax[0].legend()

ax[1].plot(wake.f*1e-9, np.abs(wake.Z), c='b', lw=1.5, label='FIT+Wakis')
ax[1].plot(cstZ[0], cstZ[1], c='k', ls='--', lw=1.5, label='CST')
ax[1].set_xlabel('f [GHz]')
ax[1].set_ylabel('Longitudinal impedance [Abs][$\Omega$]', color='b')
ax[1].legend()

#fig.suptitle('Benchmark with CST Wakefield Solver')
fig.tight_layout()
fig.savefig('results_beta1_sigma1e3/cfr_wake_Zmag.png')

plt.show()
```

```python
# CST_wake = np.loadtxt('/home/emacchia/wakis/benchmarks/betacavity/results_beta1_sigma1e3/CST_wake_z_beta1.txt')
# CST wake and impedance
cstWP = wake.read_txt('results_beta1_sigma1e3/CST_wake_z_beta1.txt')
cstZ = wake.read_txt('results_beta1_sigma1e3/CST_Z_z_beta1.txt')
cstZReIm = wake.read_txt('results_beta1_sigma1e3/CST_Z_z_ReIm_beta1.txt')


fig, ax = plt.subplots(1,2, figsize=[12,4], dpi=150)
ax[0].plot(wake.f*1e-9, np.real(wake.Z), c='r', lw=1.5, label='FIT+Wakis')
ax[0].plot(cstZReIm[0], cstZReIm[1], c='k', ls='--', lw=1.5, label='CST')
ax[0].set_xlabel('f [GHz]')
ax[0].set_ylabel('Longitudinal impedance - real part [$\Omega$]', color='r')
ax[0].legend()

ax[1].plot(wake.f*1e-9, np.imag(wake.Z), c='b', lw=1.5, label='FIT+Wakis')
ax[1].plot(cstZReIm[0], cstZReIm[2], c='k', ls='--', lw=1.5, label='CST')
ax[1].set_xlabel('f [GHz]')
ax[1].set_ylabel('Longitudinal impedance - imaginary part [$\Omega$]', color='b')
ax[1].legend()

#fig.suptitle('Benchmark with CST Wakefield Solver')
fig.tight_layout()
fig.savefig('results_beta1_sigma1e3/cfr_real_imag.png')

plt.show()
```
