```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data1 = pd.read_csv('../results_beta05/CSTwake.txt', delimiter='\t', skiprows=3, header=None, names=['x', 'y'])
data2 = pd.read_csv('CSTwake_bb.txt', delimiter='\t', skiprows=3, header=None, names=['x', 'y'])

# Interpolate data2 to match the x values of data1
interp_y2 = np.interp(data1['x'], data2['x'], data2['y'])

# Subtract the y-values
result_y = data1['y'] - interp_y2

# Create a DataFrame to store the result
result = pd.DataFrame({'x': data1['x'], 'y': result_y})

# Save the result to a new file
result.to_csv('subtracted_result.txt', sep='\t', index=False, header=False)

# Plot the original and subtracted data
plt.figure(figsize=(10, 6))
plt.plot(data1['x'], data1['y'], label='CSTwake')
plt.plot(data2['x'], data2['y'], label='CSTwake_bb')
plt.plot(data1['x'], result_y, label='Subtracted Result', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
